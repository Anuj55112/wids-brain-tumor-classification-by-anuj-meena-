```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.plot ( iris ["SepalLengthCm"] [:50], marker='o' , linestyle='-', color='b' )
plt.title("Sepal Length of the First 50 Entries")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.grid(True)
plt.show()
```

```python
iris = pd.read_csv("Iris.csv")
plt.figure(figsize=(10, 6))
colors = {"Iris-setosa": "red", "Iris-versicolor": "green", "Iris-virginica": "blue"}
for species, color in colors.items():
    subset = iris[iris["Species"] == species]
    plt.scatter(subset["SepalLengthCm"], subset["SepalWidthCm"],
                label=species, color=color, alpha=0.7, edgecolor='k')
plt.title("Scatter Plot of Sepal Length vs Sepal Width for All Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species", loc="upper right")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(iris["PetalLengthCm"], bins=20, color="purple", edgecolor="black")
plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python

plt.figure(figsize=(8, 6))
iris.boxplot(column="SepalWidthCm", by="Species", grid=False, patch_artist=True)
plt.title("Boxplot of Sepal Width Grouped by Species")
plt.suptitle("")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
mean_petal_length = iris.groupby("Species")["PetalLengthCm"].mean()
plt.figure(figsize=(8, 5))
mean_petal_length.plot(kind="bar", color=["red", "green", "blue"], edgecolor="black")
plt.title("Mean Petal Length for Each Species")
plt.xlabel("Species")
plt.ylabel("Mean Petal Length (cm)")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```
